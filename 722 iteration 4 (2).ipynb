```python
#set features
assembler = VectorAssembler(
    inputCols=["hotelIndex", "arrival_date_yearIndex","stays_in_weekend_nightsIndex","adultsIndex","childrenIndex","babiesIndex","mealIndex","market_segmentIndex","distribution_channelIndex","reserved_room_typeIndex","assigned_room_typeIndex","booking_changesIndex","deposit_typeIndex","customer_typeIndex","required_car_parking_spacesIndex","room_changeIndex"],
    outputCol="features")
```

```python
output = assembler.transform(Hotel_final)
output.printSchema()
```

```python
final_data=output.select("features",'have_requestsIndex')
```

```python
#divide train and test dataset
train_data,test_data = final_data.randomSplit([0.7,0.3])
```

```python
from pyspark.ml.classification import DecisionTreeClassifier,GBTClassifier,RandomForestClassifier
from pyspark.ml import Pipeline
```

```python
#7.2.3 Running the model
dtc = DecisionTreeClassifier(labelCol='have_requestsIndex',featuresCol='features')
rfc = RandomForestClassifier(labelCol='have_requestsIndex',featuresCol='features',numTrees=50,maxDepth=4)
gbt = GBTClassifier(labelCol='have_requestsIndex',featuresCol='features')
```

```python
dtc_model = dtc.fit(train_data)
rfc_model = rfc.fit(train_data)
gbt_model = gbt.fit(train_data)
```

```python
dtc_predictions = dtc_model.transform(test_data)
rfc_predictions = rfc_model.transform(test_data)
gbt_predictions = gbt_model.transform(test_data)
```

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator
```

```python
my_binary_eval = BinaryClassificationEvaluator(labelCol = 'have_requestsIndex')
```

```python
# the result of model
print("DTC")
print(my_binary_eval.evaluate(dtc_predictions))
```

```python
print("RFC")
print(my_binary_eval.evaluate(rfc_predictions))
```

```python
my_binary_gbt_eval = BinaryClassificationEvaluator(labelCol='have_requestsIndex', rawPredictionCol='prediction')
print("GBT")
print(my_binary_gbt_eval.evaluate(gbt_predictions))
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
```

```python
acc_evaluator = MulticlassClassificationEvaluator(labelCol="have_requestsIndex", predictionCol="prediction", metricName="accuracy")
```

```python
dtc_acc = acc_evaluator.evaluate(dtc_predictions)
rfc_acc = acc_evaluator.evaluate(rfc_predictions)
gbt_acc = acc_evaluator.evaluate(gbt_predictions)
```

```python
print("Here are the results!")
print('-'*40)
print('A single decision tree has an accuracy of: {0:2.2f}%'.format(dtc_acc*100))
print('-'*40)
print('A random forest ensemble has an accuracy of: {0:2.2f}%'.format(rfc_acc*100))
print('-'*40)
print('An ensemble using GBT has an accuracy of: {0:2.2f}%'.format(gbt_acc*100))
```
